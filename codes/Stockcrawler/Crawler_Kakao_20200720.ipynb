```python
import requests
import datetime
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
n_page= int(input("페이지 수를 입력하세요: "))
```

```python
base_url_kakao= "https://finance.naver.com/item/frgn.nhn?code=035720&page={}"

kakao_url_list= []
for i in range(n_page):
    kakao_url_list.append(base_url_kakao.format(i+1))
```

```python
kakao_resp_list= []
for url in kakao_url_list:
    kakao_resp_list.append(requests.get(url))
```

```python
kakao_soup_list= []
for resp in kakao_resp_list:
    kakao_soup_list.append(BeautifulSoup(resp.text))
```

```python
kakao_table_list= []
for soup in kakao_soup_list:
    for i in range(20): # 한 페이지에 20개의 날짜들이 있음
        kakao_table_list.append({
        'date': soup.select('table.type2')[1].select('span')[9*i].text.replace('.','-'),
        'stocks': soup.select('table.type2')[1].select('span')[9*i + 1].text.replace(',',''),
        'ratio': soup.select('table.type2')[1].select('span')[9*i + 3].text.strip().replace('%',''),
        'tradevolume': soup.select('table.type2')[1].select('span')[9*i + 4].text.replace(',',''),
        'agencyvolume': soup.select('table.type2')[1].select('span')[9*i + 5].text.replace(',',''),
        'foreignvolume': soup.select('table.type2')[1].select('span')[9*i + 6].text.replace(',',''),
        'foreignratio': soup.select('table.type2')[1].select('span')[9*i + 8].text.replace('%','')
    })
```

```python
kakao_final_list= kakao_table_list.copy()
kakao_final_list.reverse()
```

```python
kakao_final_list
```

```python
kakao_final= pd.DataFrame(kakao_final_list)
```

# type 변환하기

```python
kakao_final.date= kakao_final.date.astype('datetime64')
kakao_final.stocks= kakao_final.stocks.astype('int')
kakao_final.ratio= kakao_final.ratio.astype('float')
kakao_final.tradevolume= kakao_final.tradevolume.astype('int')
kakao_final.agencyvolume= kakao_final.agencyvolume.astype('int')
kakao_final.foreignvolume= kakao_final.foreignvolume.astype('int')
kakao_final.foreignratio= kakao_final.foreignratio.astype('float')
```
