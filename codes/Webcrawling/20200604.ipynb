```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
import matplotlib.font_manager as fm
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_family = fm.FontProperties(fname=font_path).get_name()
plt.rcParams["font.family"] = font_family
```

```python
font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
```

### 어벤져스 유튜브 크롤링

```python
driver= webdriver.Chrome('c:/chromedriver.exe')
```

```python
driver.get('https://www.youtube.com/watch?v=vW1LPbttFZ8&t=195s')
```

```python
# 스크롤 끝까지 내리기
try:
    while True:
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
except:
    print("더보기 끝")
```

```python
text_endgame= driver.find_elements_by_id('content-text')
```

```python
result_list_endgame= []

for texts in text_endgame:
    result_list_endgame.append(texts.text.strip())
```

```python
result_list_endgame
```

```python
result_str_endgame= ''.join(result_list_endgame)
```

```python
wordcloud= WordCloud(font_path=font_path,width = 2000,height = 2000)
wordcloud= wordcloud.generate_from_text(result_str_endgame)
array= wordcloud.to_array()
```

```python
plt.imshow(array,interpolation= "bilinear")
plt.show()
```

```python
df_result_list_endgame= pd.DataFrame(result_list_endgame)
```

```python
df_result_list_endgame.to_excel('endgame.xlsx')
```
