# 네이버 영화 관람객 평점_크롤링

```python
import requests
from bs4 import BeautifulSoup
```

```python
#https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=31150

base_url = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn'

params = {
    'code': '38899',
    'page': 0
}

page_limit = 0 # 뽑고자 하는 페이지 수 만큼
```

```python
result_list= []

#페이지 for
for i in range(1, page_limit+1):
    params['page'] = i
    resp = requests.get(base_url, params=params)
    soup = BeautifulSoup(resp.text, 'lxml')

#코멘트 for
    for j in range(10):
        result_list.append(soup.find('span',id='_filtered_ment_{}'.format(j)).text.strip())  
```

```python
print(result_list)
```
